First let's take a look at the report just from 2023 to see what goes into a single report, then we will later take a look at the reports made throughout the years and see how the numbers have changed over the years for some countries.

```python
df = pd.read_csv('world-happiness-report-2023.csv')
df.shape
```

```python
df['Ladder score'].describe()
```

```python
sn.histplot(df['Ladder score'])

plt.show()
```

```python
df[df['Country name'] == 'Serbia']
```

Just wanted to see where Serbia ranks in the world, from now let's explore which countries are best and worst at the specific things that go into forming a final score.

```python
df[df['Ladder score'] == df['Ladder score'].max()]
```

```python
df[df['Ladder score'] == df['Ladder score'].min()]
```

It seems that overall Finland is ranked as the happiest country while Afghanistan is the least happy country.

```python
df[df['Logged GDP per capita'] == df['Logged GDP per capita'].max()]
```

```python
df[df['Logged GDP per capita'] == df['Logged GDP per capita'].min()]
```

Luxembourg seems to be the country with the best economic production while Venezuela takes the last spot.

```python
df[df['Social support'] == df['Social support'].max()]
```

```python
df[df['Social support'] == df['Social support'].min()]
```

Afghanistan takes another last spot, this time for social support. Meanwhile the first spot belongs to another top 10 country.

```python
df[df['Healthy life expectancy'] == df['Healthy life expectancy'].max()]
```

```python
df[df['Healthy life expectancy'] == df['Healthy life expectancy'].min()]
```

Unexpectedly the country with the longest living populace isn't even close to top 10. Last place here is taken by Mozambique at 51.53 years on average.

```python
df[df['Freedom to make life choices'] == df['Freedom to make life choices'].max()]
```

```python
df[df['Freedom to make life choices'] == df['Freedom to make life choices'].min()]
```

The "freest" country is Finland offering its people the highest freedom of life choices. Afghanistan again takes another last spot.

```python
df[df['Generosity'] == df['Generosity'].max()]
```

```python
df[df['Generosity'] == df['Generosity'].min()]
```

The most generous people seem to be on average from Indonesia, again surprisingly another country that's not even close to the top 10. The last place here is taken by Georgia.
